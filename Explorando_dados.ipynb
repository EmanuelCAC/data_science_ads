```python
dados = {
    "jogador": ["Marcelo", "Pedro", "Marcelo", "Adriano", "Mauro", "Pedro", "Marcelo"],
    "infracao": ["FALTA VIOLENTA", "RECLAMACAO", "FALTA COMUM", "RECLAMACAO", "FALTA COMUM", "FALTA VIOLENTA", "RECLAMACAO"],
    "juiz_A": [4,1,3,2,4,4,2],
    "juiz_B": [2,1,4,1,1,5,6]
}

df = pd.DataFrame(dados)

print("Juiz A")
print('---------------------')
print("Amplitude: ", df["juiz_A"].max() - df["juiz_A"].min())
print("Variância: ", f"{df['juiz_A'].var():.3f}")
print("Desvio Padrão", f"{df['juiz_A'].std():.3f}")
print("\n")
print("Juiz B")
print('---------------------')
print("Amplitude: ", df["juiz_B"].max() - df["juiz_B"].min())
print("Variância: ", f"{df['juiz_B'].var():.3f}")
print("Desvio Padrão", f"{df['juiz_B'].std():.3f}")
```

```python
ja_amplitude = round(df["juiz_A"].max() - df["juiz_A"].min(), 3)
ja_variancia = round(df['juiz_A'].var(), 3)
ja_desvio = round(df['juiz_A'].std(), 3)
ja_media = round(df['juiz_A'].mean(), 3)
ja_mediana = round(df['juiz_A'].median(), 3)
ja_moda = df['juiz_A'].mode().values

jb_amplitude = round(df["juiz_B"].max() - df["juiz_B"].min(), 3)
jb_variancia = round(df['juiz_B'].var(), 3)
jb_desvio = round(df['juiz_B'].std(), 3)
jb_media = round(df['juiz_B'].mean(), 3)
jb_mediana = round(df['juiz_B'].median(), 3)
jb_moda = df['juiz_B'].mode().values

print("     Valores     |      Juiz A     |      Juiz B      ")
print("------------------------------------------------------")
print(f"{'Media':<17}|{ja_media:^17}|{jb_media:^17}")
print(f"{'Mediana':<17}|{ja_mediana:^17}|{jb_mediana:^17}")
print(f"{'Moda':<17}|{ja_moda[0]:^17}|{jb_moda[0]:^17}")
print(f"{'Amplitude':<17}|{ja_amplitude:^17}|{jb_amplitude:^17}")
print(f"{'Variância':<17}|{ja_variancia:^17}|{jb_variancia:^17}")
print(f"{'Desvio Padrão':<17}|{ja_desvio:^17}|{jb_desvio:^17}")
```

```python
notas = pd.DataFrame({
    "A1": [9.8,7.2,8.0],
    "A2": [5.3,4.0,3.5],
    "A3": [5.5,8.1,7.2],
    "A4": [7.0,7.5,6.5]
    },
    index=["P1", "P2", "P3"]
)
print("Notas Finais: ")
print("--------------------------")
print(notas)

print('\nMedia de cada aluno')
print('--------------------------')
print(notas.mean())

print('\nMaior nota de cada aluno')
print('--------------------------')
print(notas.max())

print('\nMédia de cada prova')
print('--------------------------')
print(notas.mean(axis=1))

print('\nMaior nota de cada prova')
print('--------------------------')
print(notas.max(axis=1))

print('\nMenor nota de cada prova')
print('--------------------------')
print(notas.min(axis=1))
```

```python
dados = {"nadador": ["Simonas Bills",
                     "Benjamin Proud",
                     "Antoby Ervin",
                     "Florent Manaudou",
                     "Andriy Hovorov",
                     "Nathan Adrian",
                     "Bruno Fratus",
                     "Brad Tandy"],
         "nacionalidade": ["Lituânia",
                           "Grã-Bretanha",
                           "Estados Unidos",
                           "França",
                           "Ucrânia",
                           "Estados Unidos",
                           "Brasil",
                           "Africa do Sul"],
         "tempo": [22.08,21.68,21.40,21.41,21.74,21.49,21.79,21.79]
}

raias = list(range(1,9))
prova50m = pd.DataFrame(dados, index=raias)
prova50m.index.name = "raia"

prova50m = prova50m.sort_values(by="tempo")
print("* * Resultado Final Ordenado por Tempo * *")
print(prova50m)

resultado_por_raia = prova50m['tempo'].rank(method='min')
print("\n* * Classificação por Raia * *")
print(resultado_por_raia)
```

```python
import pandas as pd

dados = {
  "sexo": ["F", "M", "F", "F", "F", "M"],
  "bairro": ["Belverde", "Belverde", "Savassi", "Anchiea", None, "Savassi"],
  "valor": [150.00, 35.00, 80.00, 250.00, 9.90, 25.00],
  "cartao": ["Master", "Visa", "Visa", "Amex", "Elo", "Master"]
}

id_clientes = [1, 2, 3, 4, 5, 6]

vendas = pd.DataFrame(dados, index=id_clientes)
```

```python
print("Dominio dos atributos categóricos")
print("-----------------------------")
print("sexo: ", vendas['sexo'].unique())
print("bairro: ", vendas['bairro'].unique())
print("cartao: ", vendas['cartao'].unique())
```

```python
print("Tabela de frequência")
print("--------------------")
print("\n1 - sexo")
print(vendas['sexo'].value_counts())
print("\n2 - bairro")
print(vendas['bairro'].value_counts())
print("\n3 - cartao")
print(vendas['cartao'].value_counts())
```

```python
group_valor_bairro = vendas['valor'].groupby(vendas['bairro'])

group_valor_bairro = vendas.groupby('bairro')['valor']

print(group_valor_bairro.count())
print(group_valor_bairro.sum())
print(group_valor_bairro.mean())
```

```python
# Estudando a base de dados Flags
import pandas as pd

# Importando os dados
flags = pd.read_csv('/content/drive/MyDrive/Faculdade/6Semestre/CienciadeDados/MaterialApoio/flags.csv')

flags
```

```python
# Obtendo as propriedade básicas de cada atributo
i = 0
for c in flags.columns:
  i += 1
  att = flags[c]                              # atributo
  att_dtype = att.dtype                       # dtype
  att_tam_dominio = att.unique().size         # tamanho do dominio
  att_tem_nulo = any(att.isnull())            # tem nulo?
  if(att_tam_dominio < 8):
    print("("+str(i)+") atributo: ", c, "\t",
          "dtype: ", att_dtype, "\t",
          "nulos: ", att_tem_nulo, "\n",
          "dominio: ", att.unique())
  else:
    if(att_dtype == 'object'):
      print("("+str(i)+") atributo: ", c, "\t",
            "dtype: ", att_dtype, "\t",
            "nulos: ", att_tem_nulo, "\n",
            "dominio (primeiros elementos): ", att.unique()[0:8])
    else:
      print("("+str(i)+") atributo: ", c, "\t",
            "dtype: ", att_dtype, "\t",
            "nulos: ", att_tem_nulo, "\n",
            "min: ", att.min(), "\t",
            "max: ", att.max(), "\t",
            "media: ", round(att.mean(),2), "\t",
            "d.p.: ", round(att.std(),2))
  print()
```

```python
df_cores = pd.DataFrame()
for c in flags.columns:
  if c in ['red', 'green', 'blue', 'gold', 'white', 'black', 'orange']:
    df_cores[c] = flags[c].value_counts()
    # df_cores.index.name = c

df_cores
```

```python
lst_cores = ['red', 'green', 'blue', 'gold', 'whitesmoke', 'black', 'orange']

df_cores.plot(kind='barh', subplots=True, figsize=(8,25), color=lst_cores)
```
